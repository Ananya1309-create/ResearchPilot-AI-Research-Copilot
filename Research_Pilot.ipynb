```python
def ask_paper(question):
    """
    Ask questions about the uploaded research paper.
    Uses ChromaDB + Gemini (RAG pipeline).
    """

    # Convert question to embedding
    query_embedding = model.encode(question)

    # Retrieve top 3 relevant chunks
    results = collection.query(
        query_embeddings=[query_embedding.tolist()],
        n_results=3
    )

    # Combine retrieved chunks
    context = "\n\n".join(results["documents"][0])

    # Prompt for Gemini
    prompt = f"""
You are an expert AI research assistant.

Answer ONLY using the context provided below.
Do not make assumptions or use outside knowledge.

If the answer is not present in the context, say:
"I could not find this information in the uploaded paper."

Context:
{context}

Question:
{question}

Answer:
"""

    # Generate response
    response = model_gemini.generate_content(prompt)

    return response.text
```

```python
from pypdf import PdfReader
from langchain_text_splitters import RecursiveCharacterTextSplitter

all_chunks = []
paper_names = []

splitter = RecursiveCharacterTextSplitter(
    chunk_size=1000,
    chunk_overlap=200
)

for pdf_file in uploaded.keys():

    print("Processing:", pdf_file)

    reader = PdfReader(pdf_file)

    text = ""

    for page in reader.pages:
        page_text = page.extract_text()

        if page_text:
            text += page_text

    chunks = splitter.split_text(text)

    for chunk in chunks:
        all_chunks.append(chunk)
        paper_names.append(pdf_file)

    print("Chunks from this paper:", len(chunks))

print("\nTotal Chunks:", len(all_chunks))
print("Total Papers:", len(set(paper_names)))
```

```python
embeddings = model.encode(all_chunks)

print("Embeddings created:", len(embeddings))
```

```python
import chromadb

client = chromadb.Client()

collection = client.create_collection(
    name="multi_paper_db"
)

ids = [str(i) for i in range(len(all_chunks))]

collection.add(
    documents=all_chunks,
    embeddings=embeddings.tolist(),
    ids=ids,
    metadatas=[
        {"paper": paper}
        for paper in paper_names
    ]
)

print("Database created successfully!")
print("Documents stored:", collection.count())
```

```python
def compare_papers(question):

    query_embedding = model.encode(question)

    results = collection.query(
        query_embeddings=[query_embedding.tolist()],
        n_results=15
    )

    contexts = []

    for doc, meta in zip(
        results["documents"][0],
        results["metadatas"][0]
    ):

        contexts.append(
            f"Paper: {meta['paper']}\n{doc}"
        )

    context = "\n\n".join(contexts)

    prompt = f"""
You are an expert research assistant.

Compare the papers using ONLY the provided context.

Generate a markdown table containing:

Paper Name
Methodology
Dataset
Key Contribution

If information is unavailable, write "Not Mentioned".

Context:
{context}

Question:
{question}
"""

    response = model_gemini.generate_content(prompt)

    return response.text
```

```python
print(compare_papers(
    "Compare the methodologies and datasets used in these papers."
))
```

```python
print(compare_papers(
    "What research gaps remain across these studies?"
))
```

```python
print(compare_papers(
    "Which paper achieved the best performance and why?"
))
```
